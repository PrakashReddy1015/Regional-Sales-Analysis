## Key Insights from Exploratory Data Analysis

- The business operates at a large scale, generating hundreds of millions in revenue.
- Monthly sales exhibit a stable trend with occasional fluctuations, indicating consistent demand.
- The Wholesale channel is the dominant contributor to both revenue and profit.
- Export channel contributes the least, suggesting potential growth opportunities.
- Sales are concentrated among a limited number of products, with some high-revenue products showing lower profit margins.



```python
#products with high revenue and low profit
high_revenue_low_profit = (
    df.groupby('product_description_index')[['revenue', 'profit']]
    .sum()
    .assign(
        revenue_m=lambda x: x['revenue'] / 1e6,
        profit_m=lambda x: x['profit'] / 1e6,
        profit_margin=lambda x: x['profit'] / x['revenue']
    )
    [['revenue_m', 'profit_m', 'profit_margin']]   #  KEEP ONLY THESE
    .sort_values(['revenue_m', 'profit_margin'], ascending=[False, True])
    .round({'revenue_m': 1, 'profit_m': 1, 'profit_margin': 3})
)

high_revenue_low_profit.head(10)
```

### Product-Level Insights

- A small number of products contribute disproportionately to total revenue.
- Some high-revenue products operate at lower profit margins, indicating potential pricing or cost optimization opportunities.
