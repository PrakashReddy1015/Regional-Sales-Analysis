```python
import pandas as pd
import numpy as np


```

```python
df = pd.read_excel("../data/raw/Regional Sales Dataset.xlsx")
```

```python
df.head()
```

```python
df.shape
df.info()
```

```python
df['OrderDate'] = pd.to_datetime(df['OrderDate'])
```

```python
df.columns = (
    df.columns
    .str.strip()
    .str.lower()
    .str.replace(' ', '_')
)
df = df.rename(columns={
    'ordernumber': 'order_number',
    'orderdate': 'order_date'
})

```

```python
df.columns
```

```python
df.isnull().sum()
```

```python
df.duplicated().sum()
```

Create new metrics

```python
df['revenue'] = df['line_total']
```

```python
df['total_cost'] = df['total_unit_cost'] * df['order_quantity']
```

```python
df['profit'] = df['revenue'] - df['total_cost']
```

```python
df['profit_margin'] = df['profit'] / df['revenue']
```

```python
##validate the new metrics
df[['revenue', 'total_cost', 'profit', 'profit_margin']].describe()
```

```python
df['year'] = df['order_date'].dt.year
df['month'] = df['order_date'].dt.month
df['month_name'] = df['order_date'].dt.month_name()
```

### Data Cleaning & Feature Engineering Summary

- Converted date columns to datetime format to enable accurate time-series analysis.
- Standardized column names using snake_case for consistency and ease of use.
- Retained repeated order_number values as they represent valid sales events rather than data quality issues.
- Created derived business metrics including revenue, total cost, profit, and profit margin.
- Extracted time-based features such as year, month, and month_name to support trend and seasonality analysis.
